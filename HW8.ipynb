```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/MSSP-607/Modules/week8/InternetSales.csv', index_col=0, parse_dates=True)
```

```python
data.columns
```

```python
# change the OrderDate to time
data["TradeDate"] = pd.to_datetime(data.OrderDate)
```

```python
# subset
df = data[["TradeDate", "SalesAmount"]]
```

```python
# get the daily sales
date_list = df.TradeDate.unique()
sales_list = []
for date in date_list:
    sales = df[df['TradeDate'] == date].SalesAmount.sum()
    sales_list.append(sales)

df2 = pd.DataFrame(list(zip(date_list, sales_list)), columns =['date', 'sales']) 

df2 = df2.sort_values(by = 'date')
```

```python
# line
fig = px.line(df2,line_shape='hvh',x='date',y='sales')
fig.show()
```

```python
# trendline 1
fig = px.scatter(df2, x='date', y='sales',trendline="lowess", color="sales")
fig.show()
```

```python
# trendline 2
fig = px.scatter(df2, x='date', y='sales', trendline="ols", color="sales")
fig.show()
```
